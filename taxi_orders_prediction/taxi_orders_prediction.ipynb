## Подготовка

Импортируем необходимые библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from lightgbm import LGBMRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

Откроем файл с данными и изучим его. Сразу же в качестве индекса укажем первый столбец, а также преобразуем этот столбец в датасете в тип datetime. 

```python
data = pd.read_csv('/datasets/taxi.csv', index_col=[0], parse_dates=[0])
#отсортируем индексы по порядку и проверим сортировку
data = data.sort_index()
display(data.index.is_monotonic)
display(data.head())
data.info()
```

Данные состоят из 2х столбцов, 26496 строк, пропуски отсутствуют. Индексы отсортированы по порядку. Посмотрим, за какой период представлены эти данные.

```python
display(data.index.min())
display(data.index.max())
```

Видим, что данные в таблице содержатся за период с марта по август 2018 года. В таблице записаны данные с интервалом в 10 минут, такие интервалы кажутся слишком частыми, выполним их ресемплирование по одному часу.

```python
data = data.resample('1H').sum() 
data.plot()
plt.title('Количество заказов такси в разрезе часа')
plt.ylabel('Количество заказов')
plt.xlabel('Время');
```

Вывод: Данные изучены, они состоят из двух столбцов - дата и время (преобразованы в индексы) и количество заказов. В данных не обнаружено пропусков. Информация о наблюдениях содержится за период с марта по август 2018 года. Также данные были ресемплированы по одному часу. Также видно увеличение заказов с теченим времени.

Прежде чем переходить к прогнозированию временного ряда, нужно , чтобы ряд стал более стационарным. Применим метод Shift() и найдем разности значений временного ряда. Добавим аргумент fill_value=0, чтобы заполнить недостающие значения на 0.

```python
data_shifted = data - data.shift(fill_value=0)
data_shifted['mean'] = data_shifted['num_orders'].rolling(24).mean()
data_shifted['std'] = data_shifted['num_orders'].rolling(24).std()
data_shifted.plot()
```

Видно, что ряд стал более стационарным, такие данные можно прогнозировать, к чему мы приступим в следующем блоке проекта.

## Обучение

### Подготовка признаков

Прежде чем приступить к обучению модели, нужно добавить в датасет признаки, на которые модель будет опираться при обучении: 
1. Календарные признаки(день недели и час)
2. «Отстающие значения» (они подскажут количество заказов за несколько предыдущих часов)
3. Скользящее среднее (задаст общий тренд, подскажет количество заказов такси за нескольько ближайщих часов)
Для этого напишем функцию make_features()

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    for rolling in range(1, rolling_mean_size + 1):
        data['rolling_mean'] = data['num_orders'].shift().rolling(rolling).mean()
```

Применим функцию к нашим данным

```python
make_features(data, 24, 3)
data.head()
```

Удалим пропущенные значения 

```python
data.dropna(inplace=True)
display(data.head())
data.info()
```

Разделим данные на обучающую и тестовую выборки

```python
train, test = train_test_split(data, shuffle=False, test_size=0.1)

features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']

features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']
```

```python
display(features_train.shape)
display(target_train.shape)
display(features_test.shape)
display(target_test.shape)
```

Теперь необходимо масштабировать признаки, чтобы модель обучалась корректно. Для этого используем Standard Scaler.

```python
scaler = StandardScaler()
scaler.fit(features_train)
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)
```

После того, как масштабировали признаки, можно приступать к обучению моделей. Для чистоты эксперимента, создадим модель, которая будет делать константное предсказание - среднее количество заказов. Вычислим ее метрику качества RMSE. Это даст нам возможность более объективно сравнить качество моделей.

### Dummy модель

```python
predictions = np.ones(len(target_test)) * data['num_orders'].mean()
rmse = mean_squared_error(target_test, predictions) ** 0.5 
print('RMSE константной модели:', rmse)
```

```python
predictions = target_test.shift(1, fill_value=target_train.iloc[-1])
rmse = mean_squared_error(target_test, predictions) ** 0.5 
print('RMSE константной модели:', rmse)
```

### Модель LightGBM

```python
model_lgbm = LGBMRegressor(random_state=12345)

# Подготовим параметры для нашей модели 
params = {'n_estimators': range (50, 100, 10),
          'max_depth': range (5, 15, 5),
          'learning_rate': [0.05, 0.1, 0.3]
          }


# Для подбора наилучших параметров будем использовать GridSearchCV 
tscv = TimeSeriesSplit(n_splits=5)
grid = GridSearchCV(estimator=model_lgbm, param_grid=params, cv=tscv, scoring = 'neg_root_mean_squared_error')

grid.fit(features_train, target_train)

print("Наилучшие параметры:", grid.best_params_)
print("RMSE:", grid.best_score_ * -1)
```

### Модель Линейной регресии

```python
# Проверим качество модели на кросс-валидации 
model = LinearRegression()

tscv = TimeSeriesSplit(n_splits=5)

cross_val_lr = cross_val_score(model, 
                             features_train, 
                             target_train, 
                             cv=tscv, 
                             scoring='neg_root_mean_squared_error')

rmse = cross_val_lr.mean() * -1
print("RMSE LinearRegression:", rmse)
```

### Модель Случайного леса

```python
model_forest = RandomForestRegressor(random_state=12345)
params_forest = {'n_estimators': range (50, 100, 10),
          'max_depth': range (5, 15, 5),
          'min_samples_split': range (2,10,2)
          }

grid = GridSearchCV(estimator=model_forest, param_grid=params_forest, cv=tscv, scoring = 'neg_root_mean_squared_error')

grid.fit(features_train, target_train)

print("Наилучшие параметры:", grid.best_params_)
print("RMSE:", grid.best_score_ *-1)
```

### Вывод: 
В результате все модели показали значительно лучшее качество по сравнению с константной моделью. Наиболее хорошее качество на кросс-валидации показала себя модель Модель LightGBM с параметрами n_estimators=90, max_depth=10, learning_rate=0.05. Ее RMSE на кросс-валидации составила 24.9

## Тестирование

Протестируем Модель LightGBM с параметрами n_estimators=90, max_depth=10, learning_rate=0.05 на тестовой выборке и оценим ее качество.

```python
model_lgbm = LGBMRegressor(random_state=12345,
                          n_estimators=90,
                          max_depth=10,
                          learning_rate=0.05)
model_lgbm.fit(features_train, target_train)

predictions = model_lgbm.predict(features_test)


rmse = mean_squared_error(target_test, predictions)**0.5
print("Итоговый RMSE", rmse)
```

Вывод: в результате тестирования модель показала RMSE 41.6, что укладывается в требования заказчика (меньше 48)

#### Этап 2. Обучение моделей
Для обучения моделей в датасет были добавлены новые признаки - Календарные (месяц, день, день недели и час), «Отстающие значения» и Скользящее среднее. Также признаки были масштабированны, чтобы не вводить в заблуждение модель.

Были обучены три модели - LightGBM, Линейную регрессию и Модель случайного леса - и были найдены наиболее подходящие гиперпараметры, благодаря которым модели выдавали наилучшее качество.

#### Этап 3. Анализ моделей
В ходе этого этапа был произведен анализ качества моделей на кросс-валидации. В результате наиболеекачественной показала себя модель LightGBM - ее RMSE составило - 24.9.

#### Этап 4. Тестирование
Модель LightGBM с наилучшими параметрами была протестирована, и показала значение RMSE 41.6, что удоблетворяет условиям заказчика.
